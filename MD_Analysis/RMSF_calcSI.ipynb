```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
rmsf_files = {
    "rmsf_all_DLC100ns.dat": "DNA-Amino.linker-CNT RMSF",
    "rmsf_DNA_DLC100ns.dat": "DNA Alone RMSF",
    "rmsf_CNT_DLC100ns.dat": "CNT Alone RMSF",
    "rmsf_Linker_DLC100ns.dat": "Linker Alone RMSF",
}

# Define colors for each dataset
colors = {
    "DNA-Amino.linker-CNT RMSF": "blue",
    "DNA Alone RMSF": "crimson",
    "CNT Alone RMSF": "darkgreen",
    "Linker Alone RMSF": "purple",
}
```

```python
valid_files = {file: title for file, title in rmsf_files.items() if os.path.exists(file)}
if not valid_files:
    print("No valid RMSF files found!")
else:
    print(f"Loaded {len(valid_files)} RMSF files.")
```

```python
rmsf_data = {}

for file, title in valid_files.items():
    data = np.loadtxt(file)
    rmsf_data[title] = data

print("RMSF data processed successfully.")
```

```python
for title, data in rmsf_data.items():
    plt.figure(figsize=(9, 6))
    plt.plot(data[:, 0], data[:, 1], linestyle='-', color=colors[title], label=title)
    plt.xlabel("Residue Index")
    plt.ylabel("RMSF (Å)")
    #plt.ylim(0, 8.5)  
    plt.title(title)
    plt.legend()
    plt.grid(True, linestyle="--", alpha=0.6)
    plt.show()
```
